```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv("train.csv")
```

```python
print(titanic_data.head())
```

```python
# Checking  missing values
print(titanic_data.isnull().sum())
```

```python
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Survived
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=titanic_data)
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Pclass
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Pclass Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Sex
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=titanic_data)
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Survival rate by Sex
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival rate by Pclass
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival rate by Age
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Survival rate by Embarked
plt.figure(figsize=(10, 6))
sns.barplot(x='Embarked', y='Survived', data=titanic_data)
plt.title('Survival Rate by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_cols = titanic_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = titanic_data[numeric_cols].corr()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
